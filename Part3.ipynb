```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data_age = pd.read_csv('Crude suicide rates.csv')
data_age
```

```python
data_age.replace([" Both sexes", " Male", " Female"], ["Both sexes", "Male", "Female"], inplace=True)
data_age
```

```python
index_sex = data_age[(data_age['Sex'] == 'Both sexes')].index
data_age.drop(index_sex, inplace=True)
data_age
```

```python
data_age['Mean'] = (
        data_age[' 80_above'] + data_age[' 70to79'] + data_age[' 60to69 '] + data_age[' 50to59 '] +
        data_age[' 40to49'] + data_age[' 30to39'] + data_age[' 20to29'] + data_age[' 10to19'])/8
data_age
```

```python
data_age_sort_mean = data_age.sort_values(by=['Mean'], ascending=False).head(10)
data_age_sort_mean
```

```python
data_age_des = data_age.describe()
data_age_des 
```

```python
data_age_sort_mean.loc['Global Mean'] = data_age_des.loc['mean']
data_age_sort_mean
```

```python
data_age_sort_mean.replace(np.nan, 'Mean of Both sexes', inplace=True)
data_age_sort_mean
```

```python
sns.stripplot(x='Country', y='Mean', data=data_age_sort_mean, size=8, hue="Sex")
plt.xticks(rotation=90)
plt.legend()
plt.grid()
plt.xlabel('Countries')
plt.ylabel('Mean of suicide rates per 100000 population')
plt.title('Which gender in which countries had the highest suicide rate in 2016')
plt.show()
```

```python
data_age_female = data_age[data_age['Sex'] == 'Female']
data_age_female
```

```python
data_age_female_des = data_age_female.describe()
data_age_female_des
```

```python
data_age_female_top10 = data_age_female.sort_values(by=['Mean'], ascending=False).head(10)
data_age_female_top10
```

```python
data_age_female_top10.loc['Global Females Mean'] = data_age_female_des.loc['mean']
data_age_female_top10
```

```python
data_age_female_top10.replace(np.nan, 'Global Females Mean', inplace=True)
data_age_female_top10
```

```python
sns.stripplot(x='Country', y='Mean', data=data_age_female_top10, size=8)
plt.xticks(rotation=90)
plt.grid()
plt.xlabel('Countries')
plt.ylabel('Women suicide rates per 100000 population')
plt.title('Which countries had the highest women suicide rate in 2016')
plt.show()
```
