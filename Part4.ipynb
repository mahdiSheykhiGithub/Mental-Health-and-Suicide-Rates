```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data_fac=pd.read_csv('Facilities.csv')
data_fac
```

```python
data_fac.dtypes
```

```python
data_fac.isnull().sum()
```

```python
data_fac.drop(['residential_facilities','day _treatment'],axis=1,inplace=True)
data_fac
```

```python
year=set(data_fac.Year)
year
```

```python
data_fac.drop('Year',axis=1,inplace=True)
data_fac.head(10)
```

```python
data_fac.drop_duplicates(inplace=True)
data_fac.shape
```

```python
data_fac.isnull().sum()
```

```python
data_fac.sort_values(by='health_units',inplace=True,ascending=False)
data_fac.tail(30)
```

```python
data_fac.interpolate(method='linear',inplace=True)
```

```python
data_fac.isnull().sum()
```

```python
data_fac
```

```python
data_age = pd.read_csv('Crude suicide rates.csv')
data_age.replace([" Both sexes", " Male", " Female"], ["Both sexes", "Male", "Female"], inplace=True)
both_sexes = data_age[data_age['Sex'] == 'Both sexes']
both_sexes.rename(index=both_sexes.Country,inplace=True)

both_sexes['Mean of all ages'] = (
        both_sexes[' 80_above'] + both_sexes[' 70to79'] + both_sexes[' 60to69 '] + both_sexes[' 50to59 '] +
        both_sexes[' 40to49'] + both_sexes[' 30to39'] + both_sexes[' 20to29'] + both_sexes[' 10to19'])/8
Suicide_Rates=both_sexes[['Country','Mean of all ages']]
Suicide_Rates
```

```python
final_data=data_fac.merge(Suicide_Rates,how='inner',on='Country')
final_data.isnull().sum()
```

```python
final_data.rename(index=final_data.Country,inplace=True)
final_data.drop('Country',axis=1,inplace=True)
final_data
```

```python
final_data.dtypes
```

```python
from sklearn.preprocessing import minmax_scale
```

```python
minmax_data=minmax_scale(final_data,feature_range=(0,5))
scale_df=pd.DataFrame(minmax_data,index=final_data.index,columns=final_data.columns)
scale_df
```

```python
final_data_top10=scale_df.sort_values(by=['Mean of all ages'],ascending=False).head(10)
final_data_top10
```

```python
final_data_des=scale_df.describe()
final_data_des
```

```python
final_data_top10.loc['Global average']=final_data_des.loc['mean']
final_data_top10
```

```python
final_data_tail10=scale_df.sort_values(by=['Mean of all ages'],ascending=False).tail(10)
final_data_tail10
```

```python
final_data_tail10.loc['Global average']=final_data_des.loc['mean']
final_data_tail10
```

```python
plt.plot(final_data_top10.index,final_data_top10['Mental _hospitals'], ls='dotted', marker='.', mew=2,label='Mental hospitals',color='navy')
plt.plot(final_data_top10.index,final_data_top10['health_units'], ls='dotted', marker='.', mew=2,label='Health units',color='red')
plt.plot(final_data_top10.index,final_data_top10['outpatient _facilities'], ls='dotted', marker='.', mew=2,label='Outpatient facilities',color='darkorange')
plt.bar(final_data_top10.index,final_data_top10['Mean of all ages'],label='Suicide Rate',alpha=0.3,color='dimgray')
plt.xticks(rotation=90)
plt.legend()
plt.grid()
plt.title('The number of health centers in the ten countries with the highest suicide rates')
plt.xlabel('Countries')
plt.show()
```

```python
plt.plot(final_data_tail10.index,final_data_tail10['Mental _hospitals'], ls='dotted', marker='.', mew=2,label='Mental hospitals',color='navy')
plt.plot(final_data_tail10.index,final_data_tail10['health_units'], ls='dotted', marker='.', mew=2,label='Health units',color='red')
plt.plot(final_data_tail10.index,final_data_tail10['outpatient _facilities'], ls='dotted', marker='.', mew=2,label='Outpatient facilities',color='darkorange')
plt.bar(final_data_tail10.index,final_data_tail10['Mean of all ages'],label='Suicide Rate',alpha=0.3,color='dimgray')
plt.xticks(rotation=90)
plt.legend()
plt.grid()
plt.title('The number of health centers in the ten countries with the lowest suicide rates')
plt.xlabel('Countries')
plt.show()
```
